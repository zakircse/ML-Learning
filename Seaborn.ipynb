Seaborn: Data Visualization Library

Importing the Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Note: Seaborn has some built-in dataset

```python
# total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# setting a theme for plots
sns.set_theme()
```

```python
# visualize the data
sns.relplot(data=tips, x = 'total_bill', y ='tip', col = 'time', hue = 'smoker', style = 'smoker', size = 'size')
```

```python
# Load the Iris dataset
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

Scatter Plot

```python
sns.scatterplot(x='sepal_length', y='petal_length', hue='species',data=iris)
```

```python
sns.scatterplot(x='sepal_width', y='petal_width', hue='species',data=iris)
```

Count Plot

Bar Chart

```python
# house price dataset
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
house = pd.DataFrame(housing.data, columns=housing.feature_names)
house['PRICE'] = housing.target
```

```python
house.head()
```

Distribution Plot

```python
sns.distplot(house['PRICE'])
```

Correlation: Possitive & Negative Correlation

```python
correlation = house.corr()
```

```python
# Constructing a Heat map
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square = True, fmt = '.1f',annot = True, annot_kws = {'size':8}, cmap='Blues')
```
