List vs Numpy Time Comparis

```python
from time import perf_counter
```

Time taken by List

```python
python_list = [i for i in range(10000)]

start_time = perf_counter()

python_list = [i+5 for i in python_list]

end_time = perf_counter()

print(end_time-start_time)
```

Time taken by Numpy

```python
np_array = np.array([i for i in range(10000)])

start_time = perf_counter()

np_array += 5

end_time = perf_counter()

print(end_time-start_time)
```

Inital Placeholders in numpy arrays

```python
# create a numpy array of zeros
x = np.zeros((4,5))
print(x)
```

```python
# create a numpy array of ones
y = np.ones((3,3))
print(y)
```

```python
# array of a particular value
z = np.full((5,4),5)
print(z)
```

```python
# create an identity matrix
i = np.eye(4)
print(i)
```

```python
# create a numpy array with random values
r = np.random.random((3,4))
print(r)
```

```python
# create integer values array within a specific range
ran = np.random.randint(10,100,(3,4))
print(ran)
```

```python
# array with evenly spaced values
d = np.linspace(10,30,5)
print(d)
```

```python
# array of evenly spaced values --> specifying the step
e = np.arange(10,30,5)
print(e)
```

```python
# convert a list to a numpy array
list2 = [10,20,30,40]

np_array = np.asarray(list2)
print(np_array)
type(np_array)
```

Mathematical operations on an np array

```python
list1 = [1,2,3,4,5]
list2 = [6,7,8,9]

print(list1+list2) #concatenate of joins two list
```

```python
a = np.random.randint(0,10,(3,3))
b = np.random.randint(10,20,(3,3))
```

```python
print(a)
print(b)
```

```python
print(a+b)
print(a-b)
print(a*b)
print(a/b)
```

```python
print(np.add(a,b))
```

```python
print(np.subtract(a,b))
```

```python
print(np.multiply(a,b))
```

```python
print(np.divide(a,b))
```

Array Manipulation

```python
array = np.random.randint(0,10,(2,3))
print(array)
print(array.shape)
```

```python
#Transpose
trans = np.transpose(array)
print(trans)
print(trans.shape)
```

```python
array = np.random.randint(20,30,(4,5))
print(array)
print(array.shape)
```

```python
trans2 = array.T
print(trans2)
print(trans2.shape)
```

```python
#reshaping an array
a = np.random.randint(0,10,(2,3))
print(a)
print(a.shape)
```

```python
b = a.reshape(3,2)
print(b)
print(b.shape)
```
